```python
import math
```

```python
num = 9
```

```python
angle_degrees = 45
```

```python
print("Square Root:", math.sqrt(num))
print("Factorial:", math.factorial(5))
print("Logarithm base 10:", math.log10(100))
print("Sine of 45 degrees:", math.sin(math.radians(angle_degrees)))
print("Value of Pi:", math.pi)
```
